```python
import numpy as np
import random
import math
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
```

```python
#Clustering
#K-means
```

```python
mean = np.array([[-1,-1],[1,-1],[0,1]])
cov = np.array([[[2,0.5],[0.5,1]], [[1,-0.5],[-0.5,2]], [[1,0],[0,2]]])
clusterLabels = ['a', 'b', 'c']
sigma = [0.5, 1, 2, 4, 8]
```

```python
dataset = []
for i in sigma:
    ds = []
    for j in range(len(mean)):
        x = [[k.tolist(), clusterLabels[j]] for k in np.random.multivariate_normal(mean[j], i * cov[j], 100)]
        ds += x
    dataset.append(ds)
```

```python
def kmeans(ds, numClusters=3, mean=mean, clusterLabels=clusterLabels):
    data = [i[0] for i in ds]
    trueLabel = [i[1] for i in ds]
    centroids = random.sample(data, k=numClusters)
    labelList = []
    while True:
        clusters = {}
        currLabelList = []
        for i in data:
            shortestDist = math.inf
            predictedCluster = None
            for j in range(len(centroids)):
                dist = math.dist(i, centroids[j])
                if dist < shortestDist:
                    shortestDist = dist
                    predictedCluster = j
            if not predictedCluster in clusters:
                clusters[predictedCluster] = []
            clusters[predictedCluster].append(i)
            currLabelList.append(predictedCluster)
        if currLabelList == labelList:
            mapping = {}
            for j in range(len(centroids)):
                distToTrueCentroids = [math.dist(centroids[j], k.tolist()) for k in mean]
                minIndex = distToTrueCentroids.index(min(distToTrueCentroids))
                mapping[j] = clusterLabels[minIndex]
            finalLabelList = [mapping[k] for k in labelList]
            objective = 0
            accuracy = 0
            for i in range(len(data)):
                objective += math.dist(data[i], centroids[labelList[i]])
                if finalLabelList[i] == trueLabel[i]:
                    accuracy += 1
            accuracy = accuracy / len(data)
            return centroids, finalLabelList, objective, accuracy
        for j in range(len(centroids)):
            if j in clusters:
                centroids[j] = np.mean(np.array(clusters[j]), axis=0).tolist()
        labelList = currLabelList
```

```python
def gmm(ds, numClusters=3, mean=mean, clusterLabels=clusterLabels):
    data = [i[0] for i in ds]
    trueLabel = [i[1] for i in ds]
    clusterMean = random.sample(data, k=numClusters)
    clusterCov = [np.eye(2).tolist()] * numClusters
    z = [(1/numClusters)] * numClusters
    labelList = []
    counter = 0
    while True:
        w = []
        wraw = []
        currLabelList = []
        for i in data:
            wi = []
            wrawi = []
            for j in range(numClusters):
                wi.append(z[j] * multivariate_normal(mean=clusterMean[j], cov=clusterCov[j]).pdf(i))
                wrawi.append(z[j] * multivariate_normal(mean=clusterMean[j], cov=clusterCov[j]).pdf(i))
            wi = np.array(wi) 
            wi = wi / np.sum(wi)
            wi = wi.tolist()
            w.append(wi)
            wraw.append(wrawi)
            currLabelList.append(wi.index(max(wi)))
        if currLabelList == labelList:
            counter += 1
        else:
            counter = 0
        if counter >= 5:
            mapping = {}
            for j in range(len(clusterMean)):
                distToTrueCentroids = [math.dist(clusterMean[j], k.tolist()) for k in mean]
                minIndex = distToTrueCentroids.index(min(distToTrueCentroids))
                mapping[j] = clusterLabels[minIndex]
            finalLabelList = [mapping[k] for k in labelList]
            objective = 0
            accuracy = 0
            for i in range(len(data)):
                objective += np.log(sum(wraw[i]))
                if finalLabelList[i] == trueLabel[i]:
                    accuracy += 1
            accuracy = accuracy / len(data)
            return clusterMean, finalLabelList, -objective, accuracy, counter
        labelList = currLabelList
        w = np.array(w)
        for i in range(numClusters):
            z[i] = np.mean(w[:,i])
            # print(w[:,i])
            clusterMean[i] = (np.sum(np.array([w[:,i][j] * np.array(data[j]) for j in range(len(data))]), axis=0)  / np.sum(w[:,i])).tolist()
            covTemp = np.zeros(np.array(clusterCov[i]).shape)
            for j in range(len(data)):
                covTemp += w[:,i][j] * (np.array([np.array(data[j]) - np.array(clusterMean[i])]).T @ np.array([np.array(data[j]) - np.array(clusterMean[i])]))
            covTemp = covTemp / np.sum(w[:,i])
            clusterCov[i] = covTemp.tolist()
```

```python
objectiveKmeans = []
objectiveGMM = []
accuracyKmeans = []
accuracyGMM = []
for i in range(len(dataset)):
    k = kmeans(dataset[i])
    g = gmm(dataset[i])
    objectiveKmeans.append(k[2])
    accuracyKmeans.append(k[3])
    objectiveGMM.append(g[2])
    accuracyGMM.append(g[3])
plt.
```

```python
plt.subplot()
plt.plot(sigma, objectiveKmeans, marker = 'o')
plt.xlabel("sigma")
plt.ylabel("objective value")
plt.title("KMeans clustering objective")
for x, y in zip(sigma, objectiveKmeans):
    plt.text(x, y, '({}, {})'.format(x, y))
plt.show()
```

```python
plt.subplot()
plt.plot(sigma, objectiveGMM, marker = 'o')
plt.xlabel("sigma")
plt.ylabel("objective value")
plt.title("GMM clustering objective")
for x, y in zip(sigma, objectiveGMM):
    plt.text(x, y, '({}, {})'.format(x, y))
plt.show()
```

```python
plt.subplot()
plt.plot(sigma, accuracyKmeans, marker = 'o')
plt.xlabel("sigma")
plt.ylabel("accuracy")
plt.title("KMeans clustering accuracy")
for x, y in zip(sigma, accuracyKmeans):
    plt.text(x, y, '({}, {})'.format(x, y))
plt.show()
```

```python
plt.subplot()
plt.plot(sigma, accuracyGMM, marker = 'o')
plt.xlabel("sigma")
plt.ylabel("accuracy")
plt.title("GMM clustering accuracy")
for x, y in zip(sigma, accuracyGMM):
    plt.text(x, y, '({}, {})'.format(x, y))
plt.show()
```

```python
#PCA
data2d = np.loadtxt("data2D.csv", delimiter=",", dtype=float)
data1000d = np.loadtxt("data1000D.csv", delimiter=",", dtype=float)
```

```python
def dro(data, d):
    b = np.mean(data, axis=0)
    data = data - b
    U, s, Vh = np.linalg.svd(data, full_matrices=True)
    S = np.diag(s[0:d])
    Z = np.sqrt(len(data)) * U[:,0:d]
    A = (1 / np.sqrt(len(data))) * Vh[0:d,:].T @ S
    vec1 = np.empty(shape=(len(data),1)) 
    vec1.fill(1)
    reconstruction = Z @ A.T + vec1 @ np.array([b])
    return Z, reconstruction, s
    
```

```python
def buggyPCA(data, d):
    U, s, Vh = np.linalg.svd(data, full_matrices=True)
    S = np.diag(s[0:d])
    V = Vh[0:d,:].T
    Z = U[:,0:d] @ S
    reconstruction = Z @ V.T
    return Z, reconstruction
```

```python
def demeanedPCA(data, d):
    mean = np.mean(data, axis=0)
    data = data - mean
    U, s, Vh = np.linalg.svd(data, full_matrices=True)
    S = np.diag(s[0:d])
    V = Vh[0:d,:].T
    Z = U[:,0:d] @ S
    vec1 = np.empty(shape=(len(data),1)) 
    vec1.fill(1)
    reconstruction = Z @ V.T + vec1 @ np.array([mean])
    return Z, reconstruction
```

```python
def normalizedPCA(data, d):
    mean = np.mean(data, axis=0)
    data = data - mean
    colStd = []
    for i in range(data.shape[1]):
        Std = np.std(data[:,i])
        data[:,i] = data[:,i] / Std
        colStd.append(Std)
    U, s, Vh = np.linalg.svd(data, full_matrices=True)
    S = np.diag(s[0:d])
    V = Vh[0:d,:].T
    Z = U[:,0:d] @ S
    reconstruction = Z @ V.T
    for i in range(data.shape[1]):
        reconstruction[:,i] = reconstruction[:,i] * colStd[i]
    vec1 = np.empty(shape=(len(data),1)) 
    vec1.fill(1)
    reconstruction = reconstruction + vec1 @ np.array([mean])
    return Z, reconstruction
```

```python
def plotPCA(pca, title):
    plt.subplot()
    plt.scatter(data2d[:,0], data2d[:,1], marker='o', facecolors='none', edgecolors='b')
    result = pca[1]
    plt.scatter(result[:,0], result[:,1], marker='x', color="r")
    plt.xlim(0,10)
    plt.ylim(0,10)
    plt.title(title)
    plt.show()
```

```python
plotPCA(buggyPCA(data2d,1), "Buggy PCA")
```

```python
plotPCA(demeanedPCA(data2d,1), "Demeaned PCA")
```

```python
plotPCA(normalizedPCA(data2d,1), "Normalized PCA")
```

```python
plotPCA(dro(data2d,1), "DRO")
```
